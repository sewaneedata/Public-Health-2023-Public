#### ICD10 Counts

```python
# top 10 ICD counts overall

def top10ICDoverall():
    patients_year = encc.groupby('patientid')['icd10'].unique().reset_index(name='all_unique_diag')

    # function that creates a dictionary between diagnoses and the amount of patients that have that diagnosis
    def count_diag(x):
        d={}
        for i in x:
            if i in d.keys():
                d[i]+=1
            else:
                d[i]=1
        return d
    # sorting the dictionary from most to least frequent diagnosis
    def value_getter(item):
        return item[1]
    lis = [a for b in patients_year['all_unique_diag'].tolist() for a in b]
    diag_top10 = count_diag(lis)
    popular_diag = sorted(diag_top10.items(), key = value_getter)[::-1]
    # defining a list of the top 10 diagnoses
    top_10_diag = popular_diag[:10]

    # making it a list of the diagnosis names
    top_10_diag = [x[0] for x in top_10_diag]
    
    # creating a dictionary between icd10 codes and their descriptions
    codes = {}
    shortened_icd = ["General Adult Exam", "Issue w/ Repeat Perscription", "Hypertension", "Hyperlipidemia", "Reflux Disease", "Mixed Hyperlipidemia", "Type 2 Diabetes", "Hypercholesterolemia", "Anxiety", "Respiratory Infection"]
    
    for diag in top_10_diag:
        codes[diag.split(': ')[0]] = diag.split(': ')[1]
    for code in range(len(codes.keys())):
        codes[list(codes.keys())[code]] = shortened_icd[code]
    
    # removing the codes from the names (or keeping just the nums depending on what you want to do)
    top_10_diag = [x.split(': ')[0] for x in top_10_diag]
    
    return top_10_diag
```

#### Age Distribution

```python
def age_dist(year):
    # getting data for that year
    cond_year = cond[cond['enc year'] == year]
    # grouping by patientid so there is one row per patient
    patients_year = cond_year.groupby('patientid').agg({
        'patientdob': 'first',
    }).reset_index()
    # calculating the age of each patient and putting it in a list
    age_years = []
    for dob in patients_year['patientdob']:
        age_years.append(int(year) - int(dob.split('/')[-1]))
    # creating the new age col with each age
    patients_year['age'] = [int(x) for x in age_years]
    # define age ranges
    age_ranges = [(18, 36), (37, 54), (55, 72)]
    # count the number of patients in each age range
    age_range_counts = [patients_year[(patients_year['age'] >= start) & (patients_year['age'] <= end)].shape[0]
                        for start, end in age_ranges]
    # create the age range labels for plotting
    age_range_labels = [f"{start}-{end}" for start, end in age_ranges]
    # create the plot
    fig = px.bar(x=age_range_labels, y=age_range_counts,
                 labels={
                     'x': "Age Range",
                     'y': "Number of Patients"
                 },
                 title=f"Age Range Distribution for the Year {year}")

    # hovertemplate
    fig.update_traces(hovertemplate='Age Range: %{x}<br>Count: %{y}')
    fig.update_layout(xaxis={'categoryorder': 'array', 'categoryarray': age_range_labels})

    return fig
```

```python
# Graph of top ten ICD-10 by population percentage and color as rural or urban

def diagnosis_percentage():
    df = filtered_tn.copy()
    # Initialize lists to store the percentages
    percentages_rural = []
    percentages_urban = []

    # Calculate the percentage for each diagnosis
    for diagnosis in df['Diag1'].value_counts().head(10).index:
        total_count = df[df['Diag1'] == diagnosis].shape[0]
        rural_count = df[(df['Diag1'] == diagnosis) & (df['Rural'] == True)].shape[0]
        rural_percentage = (rural_count / total_count) * 100
        percentages_rural.append(rural_percentage)
        percentages_urban.append(100 - rural_percentage)

    # Create the bar graph
    fig = go.Figure()
    # Add the bar for Rural
    fig.add_trace(go.Bar(x=df['Diag1'].value_counts().head(10).index, y=percentages_rural, name="Rural"))
    # Add the bar for Urban
    fig.add_trace(go.Bar(x=df['Diag1'].value_counts().head(10).index, y=percentages_urban, name="Urban"))
    # Customize the layout
    fig.update_layout(
        title="Percentage of Rural vs Urban by Diagnosis",
        xaxis_title="Diagnosis",
        yaxis_title="Percentage",
        yaxis_tickformat=".1f"
    )
    # Show the graph
    return fig
```
